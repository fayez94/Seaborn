#Seaborn Bar Plots Part 1 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names() #seaborn provide pre-processed data set.
```

```python
sns.load_dataset('tips')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='day', y='total_bill',data=tips) #here data meanse source of the data
```

#Hue attribute(split bar according to value of particular column)

```python
sns.barplot(x='day', y='total_bill',hue='sex',data=tips) 
```

#Palette attribute (using palplot() function to draw plot)

```python
#viewing some palettes (deep,light,winter_r,spring)
sns.color_palette('deep',5)
```

```python
sns.palplot(sns.color_palette('deep',10)) #with pallette function
```

```python
sns.palplot(sns.color_palette('spring',10))
```

```python
sns.palplot(sns.color_palette('winter',20))
```

#Creating Custom palette

```python
custom=['red','blue','green'] #create color as an array
sns.set_palette(custom) #set the palette as current palette
sns.palplot(sns.color_palette(custom)) #view the palette
```

#horizontal barplot

```python
sns.barplot(x='total_bill', y='day',hue='sex',data=tips) 
```

#Order attribute to change order of barplot

```python
sns.barplot(x='day', y='total_bill',data=tips) 
```

```python
sns.barplot(x='day', y='total_bill',order=['Sun','Fri','Sat','Thur'],data=tips) 
```

#Seaborn Bar Plots Part 2

```python
#estimator attribute
sns.barplot(x='day', y='total_bill',data=tips) 
```

```python
sns.barplot(x='day', y='total_bill',estimator=np.median,data=tips)    #not clear about estimator
```

```python
sns.barplot(x='day', y='total_bill',data=tips) 
#here 4 bar
#top of the bar denotes mean
#black of the bar means confidence interval or error bar
```

```python
#confidence interval/error bar
#confidence interval is a range of values in which we are fairly confident that the statistical messure lies it.

sns.barplot(x='day', y='total_bill',ci=60,data=tips) #error bars become short
```

```python
#capsize attribute 
#it put two horizontal line top of the error bars and bellow of the error bars
sns.barplot(x='day', y='total_bill',ci=60,capsize=0.2,data=tips)
```

```python
sns.barplot(x='day', y='total_bill',ci=60,capsize=0.5,data=tips)
```

```python
#Barplots of same color.
sns.barplot(x='day', y='total_bill',color='green',ci=60,capsize=0.5,data=tips) #all the bins/bars will be green.
```

```python
#Saturation value of particular bins
sns.barplot(x='day', y='total_bill',ci=60,capsize=0.5,data=tips, saturation=0.7)
```

```python
sns.barplot(x='day', y='total_bill',ci=60,capsize=0.5,data=tips, saturation=0.2) #bins will be lighter or dark after using saturation
```

#Python Seaborn Tutorials 4: (DistPlots)
#also known as distribution plot 

#loading penguin dataset

```python
sns.load_dataset('penguins')
```

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

#Drawing a simple DisPlot

```python
sns.displot(penguins.flipper_length_mm)
```

```python
sns.displot(penguins.flipper_length_mm,rug=True)       #show lines below the bins with rug.
```

#Removing KDE plot


```python
sns.displot(penguins.flipper_length_mm,hist=False)
```

#Setting the number of bins

```python
sns.displot(penguins.flipper_length_mm,rug=True,bins=20)    #number of beans are increased
```

#setting bin values as an array

```python
sns.displot(penguins.flipper_length_mm,bins=[160,180,200,220,240]) 
```

#fitting the data to a particular distribution using norm_hist

```python
sns.displot(penguins.flipper_length_mm,norm_hist=True) #it gives sometimes error. That's it is not recommended.
```

#styling Rugplots

```python
sns.displot(penguins.flipper_length_mm,rug_kws={'color': 'green','height':0.3},rug=True) #kws =keywords
```

#Styling histogram

```python
sns.displot(penguins.flipper_length_mm,kde=False,hist_kws={'color':'blue','alpha':0.3,'histtype':'step'})    #it's not clear
```

#stylin of kde

```python
sns.displot(penguins.flipper_length_mm,kde=True,hist=False)  #have a trouble
```

#Box Plots Python Seaborn Tutorials 5

```python
#categorical distribution plot
#A way of displaying the distribution of data based on summary statistics
#tells the values of outliers
#identify distributions of the data
#skewness of the data
penguins.describe() #to show the statistical of penguins table
```

#Basic Boxplot

```python
sns.boxplot(penguins.bill_length_mm) # describe bill_length_mm with Boxplot
```

```python
sns.boxplot(penguins.flipper_length_mm) # describe flipper_length_mm with Boxplot
```

#according to x and y axis.

```python
sns.boxplot(y=penguins.bill_length_mm,x=penguins.island) #in x axis there are 3 island
```

#Hue attribute as sex

```python
sns.boxplot(y=penguins.bill_length_mm,x=penguins.island,hue=penguins.sex)
```

#horizontal bars

```python
sns.boxplot(x=penguins.bill_length_mm,y=penguins.island,hue=penguins.sex)
```

#ordering bars

```python
sns.boxplot(x=penguins.bill_length_mm,y=penguins.island,hue=penguins.sex,order=['Biscoe','Torgersen','Dream'])
```

#color attributes

```python
sns.boxplot(x=penguins.bill_length_mm,y=penguins.island,hue=penguins.sex,order=['Biscoe','Torgersen','Dream'],color='g')
```

#width property

```python
sns.boxplot(x=penguins.bill_length_mm,y=penguins.island,hue=penguins.sex,order=['Biscoe','Torgersen','Dream'],color='g',width=1.5)
#it's not recommended. Seaborn itself provide width depending on data
#here overlap occured. so it's not recommended also
```

#whis propertry

```python
sns.boxplot(x=penguins.bill_length_mm,y=penguins.island,hue=penguins.sex,order=['Biscoe','Torgersen','Dream'],whis=0.5)
#outliers created and vertical line/viscars discreased
```

#showcaps property

```python
sns.boxplot(x=penguins.bill_length_mm,y=penguins.island,hue=penguins.sex,order=['Biscoe','Torgersen','Dream'],showcaps=False) #True is by default
#viscars/vertical lines disappear
```
